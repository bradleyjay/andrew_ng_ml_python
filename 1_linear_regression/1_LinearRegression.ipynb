```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

# read in dataset

np.seterr(divide='ignore', invalid='ignore')
data = np.genfromtxt('./ex1data1.txt', delimiter=',')

# x = np.transpose(np.array(data[:,0])) # returns vector with no hard
# directionality - not useful for matrix multiplication by default

x = np.transpose(np.matrix(data[:,0])) # default is row vector, want column
y = np.transpose(np.matrix(data[:,1]))
m = len(y)

# add ones, 
X = np.insert(x,0,1,axis=1)
```

### 2.1 Plotting the Data

```python
# plot raw data

fig, ax = plt.subplots(figsize=(12,8))

# using a matrix for X and Y here breaks this syntax...
# ax.scatter(X[:,1], y[:,0], marker = 'x', c='r') 

# and this keeps the line in...
# ax.plot(X[:,1], y[:,0], marker = 'x', c='r') 

# instead,
ax.plot(X[:,1], y[:,0], 'rx', markersize = 10) 
ax.grid(True)
ax.set_title('Food Truck - Estimated Revenue per City')
ax.set_xlabel('Population of City (10,000s)')
ax.set_ylabel('Profit ($10,000s)')
```
